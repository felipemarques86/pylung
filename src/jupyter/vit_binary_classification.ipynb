```python
import os
 
import sys
sys.path.append(os.getcwd() + '\\..\\')

from pylung import datasets

# FOUR datasets will be displayed
datasets('lidc_idri')
```

```python
from pylung import get_image_data


#this process might take several seconds
get_image_data('DS_Bravo', 'lidc_idri', index=37739)
```

The image on the left side is a slice of a CT scan with the bounding box, classification of malignancy from 1 to 5 and the degree of "how easy it is to classify this image". 
On the right side there is an image of the nodule isolated (cut from the left image) within the boundaries of the bounding box.

```python
from pylung import predict
import os

weights_dir = os.getcwd() + '\\..\\weights\\classification_one_hot_crop_nodule'

predict('DS_Bravo', 'lidc_idri', weights_dir + '\\vit_1677338358477463300_a0.8475279211997986', 37739)
```

The result is displayed with one-hot notation.
[0, 1] is for malignant nodule
[1, 0] is for benign nodule
